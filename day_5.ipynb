# <u>**Problem 1**</u>

### Construct a 2-D $10*10$ matrix with the following constraints:

*  ####  The element at $[0,0]$ should be 1
*  ####   The element at $[0,9]$ should be 100
* ####   The element at $[9,0]$ should be 50
* ####   The interval between elements in rows should be constant
* ####   The interval between elements in columns should be constant



```
# Define the matrix size
rows = 10
cols = 10

# Initialize the matrix with zeros
matrix = [[0 for j in range(cols)] for i in range(rows)]

# Set the specified constraints
matrix[0][0] = 1
matrix[0][9] = 100
matrix[9][0] = 50

# Calculate the constant intervals for rows and columns
row_interval = (matrix[0][9] - matrix[0][0]) / (cols - 1)
col_interval = (matrix[9][0] - matrix[0][0]) / (rows - 1)

# Fill in the matrix with the constant intervals
for i in range(rows):
    for j in range(cols):
        matrix[i][j] = matrix[0][0] + i * col_interval + j * row_interval

# Print the resulting matrix
for row in matrix:
    print(row)
```

#<u>**Problem 2**</u>

### Construct a 2-D $10*10$ matrix with the diagonal elements equal to zero and all the other elements equal to 1

```
# Your code below:       # Hint : Use nested for loops


 # Define the matrix size
rows = 10
cols = 10

# Initialize the matrix with ones
matrix = [[1 for j in range(cols)] for i in range(rows)]

# Set zeros on the diagonal
for i in range(rows):
    for j in range(cols):
        if i == j:
            matrix[i][j] = 0

# Print the resulting matrix
for row in matrix:
    print(row)

```

# <u>**Problem 3**</u>

### Construct a list of first 100 natural numbers.Create a two dimensional array with 10 rows and whose each row contains 10 columns and each number in this array is the square of the number from your constructed list.

### For example : [[1,4,9,...,100],[121,144,...],......,]

```
# Create a list of the first 100 natural numbers
first_100_natural_numbers = list(range(1, 101))

# Initialize a two-dimensional array with 10 rows and 10 columns
two_dimensional_array = []

# Populate the array with squares of the numbers
for i in range(0, 100, 10):
    row = [num ** 2 for num in first_100_natural_numbers[i:i+10]]
    two_dimensional_array.append(row)

# Print the resulting two-dimensional array
for row in two_dimensional_array:
    print(row)
```
